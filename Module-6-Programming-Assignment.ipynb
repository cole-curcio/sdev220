# 13.1

Write the current date as a string to the text file today.txt.

```python
from datetime import date
current_date = date(2026, 3, 2)
fmt = "Today's %A, %B, %d, %Y"
print(current_date.strftime(fmt))
filename = "today.txt"
with open("filename", "w", encoding="utf-8") as filename:
    filename.write(current_date.strftime(fmt))

```

# 15.1

Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between zero and one, print the current time, and then exit.

```python
import multiprocessing
import random
import time

def whoami():
    print("Current time:", time.ctime())
    
if __name__ == "__main__":
    for n in range(4):
        p = multiprocessing.Process(target=whoami,
                                    args=("I'm function %s" % n))
        p.start()
```
